# Handling Missing values and duplicates

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("bangalore.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df.shape
```

```python
df1=df.fillna(value=0)
```

```python
df1.isnull().sum().sum()
```

```python
df1
```

```python
df2=df.fillna(method='pad')
```

```python
df2
```

```python
df3=df.fillna(method='bfill')
```

```python
df3.head(6)
```

```python
df4=df.fillna(method='pad',axis=1)
```

```python
df4
```

```python
df5=df.fillna(method='bfill',axis=1)
```

```python
df5.isnull().sum().sum()
```

```python
df6=df.fillna(value=df['balcony'].min())
```

```python
df6
```

## replace missing values

```python
df7=df.replace(to_replace=np.nan, value='hanad')
```

```python
df7
```

## drop missing values

```python
df8=df.dropna()
```

```python
df8.shape
```

```python
df.shape
```

```python
df9=df.dropna(how='all')
```

```python
df9.shape
```
